```python
import pandas as pd

df = pd.read_csv("finalout.csv", sep=' ', dtype=str)
```

```python
df.head()
df['dash'] = '-'
df['space'] = ' '
```

```python
df['arrival'] = df['Ayear']+df['dash']+df['Amonth']+df['dash']+df['Aday']+df['space']+df['Atime']
df['closure'] = df['Cyear']+df['dash']+df['Cmonth']+df['dash']+df['Cday']+df['space']+df['Ctime']
df.arrival = pd.to_datetime(df.arrival)
df.closure = pd.to_datetime(df.closure)
df.head()
```

```python
df['time'] = (df.closure-df.arrival).astype('timedelta64[h]')
```

```python
df1 = df[['time','post', 'Amonth']]
df1.head()
```

```python
df1 = df1[df1.post != "00083"]
df1 = df1[df1.post != "11695"]
df1 = df1[df1.post != "12345"]
```

```python
table = pd.pivot_table(df1, index='post',columns='Amonth', values='time')
table
```

```python
y = [table['01'].sum(), table['02'].sum(), table['03'].sum(), table['04'].sum(), table['05'].sum(), table['06'].sum(), table['07'].sum(), table['08'].sum(), table['09'].sum()]
y
```
